### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
dataframe = pd.read_csv(url, sep='|')
```

### Step 3. Assign it to a variable called users.

```python
users = dataframe
```

```python
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby(by='occupation').age.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
def format(x):
  if x == 'M':
    return 1
  if x == 'F':
    return 0
```

```python
users['gender_numerical'] = users['gender'].apply(format)
users.head()
```

```python
male_counts = users.groupby(by= 'occupation').gender_numerical.sum()
male_counts
```

```python
male_female = users.groupby(by= 'occupation').gender.describe()['count']
male_female
```

```python
ratio = male_counts / male_female
ratio * 100
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation').age.agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(by=['occupation', 'gender']).age.mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
male_ratio = male_counts / male_female
male_ratio * 100
```

```python
female_ratio = 1 - (male_counts / male_female)
female_ratio * 100
```

我这种方法虽然是对的，但是它没有把所有的ratio展现在同一个页面，看看别人做的↓

```python
# 创建一个dataframe，包含gender和gender counts
gender_ocup = users.groupby(['occupation', 'gender']).agg({'gender': 'count'}) #注意这个count是agg里面gender的属性，例如min max这些，由于这个不是numerical的column所以没有
gender_ocup
```

```python
# 创建一个计算occupation总数的DF
occup_count = users.groupby(['occupation']).agg('count')
occup_count
```

```python
# 将分男女的除以总数
occup_gender_ratio = gender_ocup.div(occup_count, level = "occupation") * 100  #这个level估计就是指明除以哪，不太懂，之后再学
occup_gender_ratio
```
