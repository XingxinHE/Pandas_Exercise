### Step 1. Import the necessary libraries

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv). 

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv'
dataframe = pd.read_csv(url)
```

### Step 7. View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline

```python
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']] #subset部分columns常用到的syntax是这个
discipline.head()
```

### Step 8. Sort the teams by Red Cards, then to Yellow Cards

```python
discipline = discipline.sort_values(by=['Red Cards', 'Yellow Cards'])
discipline
```

### Step 9. Calculate the mean Yellow Cards given per Team

```python
round(discipline['Yellow Cards'].mean())
```

### Step 11. Select the teams that start with G

#### 我的做法如下
做的时候就在想有没有更简单的方法

```python
team_G = []
for team in euro12.loc[:,'Team'].values:
  if team[0] == 'G':
    print(team)
    team_G.append(team)
  else:
    continue
team_G
```

更简单的做法：

```python
mask_G = euro12['Team'].str.startswith('G') #这个厉害，.str转为string然后用startswith
euro12[mask_G]
```

### Step 14. Present only the Shooting Accuracy from England, Italy and Russia

#### 我的做法
的确有更简单的做法

```python
shooting_acc = euro12.loc[:,['Team','Shooting Accuracy']]
shooting_acc
```

```python
shooting_acc = shooting_acc.loc[(shooting_acc['Team'] == 'England') | (shooting_acc['Team'] == 'Italy') | (shooting_acc['Team'] == 'Russia')]
shooting_acc
```

更简单的做法：

```python
euro12.loc[euro12.Team.isin(['England', 'Italy',' Russia']),['Team', 'Shooting Accuracy']]
```

```python
euro12.loc[euro12['Team'].isin(['England', 'Italy',' Russia']),['Team', 'Shooting Accuracy']]
```

这种syntax的格式是：
dataframe.loc[ boolean mask for rows, [list of columns] ]
